```python
import numpy as np
#matrix=zeros((1024,1024))
def elipsoid (major,minor,thet,x0,y0):
    dummymat=(zeros((1024,1024)))
    matrix=zeros((1024,1024))
    for i in range(int(x0-major-1),int(x0+major+1)):
        for j in range(int(y0-major-1),int(y0+major+1)):
            if i<1024-major and i>0 and j<1024-major and j>0:
                if (i-x0)**2/(major+5)**2+(j-y0)**2/(minor+5)**2<=1:
                    dummymat[i,j]=1
                    if i<1024 and i>0 and j<1024 and j>0:
                        a=np.cos(thet/3.14*180.)*abs(i-x0)+y0
                        b=np.sin(thet/3.14*180.)*abs(j-y0)+x0
                        if a>0 and b>0 and a<1024 and b<1024:
                            matrix[int(b),int(a)]=dummymat[i,j]
                            matrix=array(matrix)
    return matrix
```
